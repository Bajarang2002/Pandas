```python
import pandas as pd
import numpy as np
df = pd.read_csv("Top100moststreamed.csv")
df.head(10)
```

```python
top10_gen =df['top genre'].value_counts()
top10_gen =top10_gen.sort_values(ascending=False).head()
top10_gen
```

```python
artist_req =['Post Malone','Ed Sheeram','Imagine Dragons','The Weekend','The Chainsmokers']
artist_req
```

```python
top10_gen.index
```

```python
r4 = df[df['top genre'].isin(top10_gen.index)].groupby('top genre').agg(CountLength =('length','count'))
r4
```

```python
r4['MaxRank'] = r4['CountLength'].rank(method = 'max')
r4['MinRank'] = r4['CountLength'].rank(method = 'min')
r4['MeanRank'] = r4['CountLength'].rank(method = 'average')
r4
```

```python
r4['MaxRank_desc']=r4['CountLength'].rank(method ='max',ascending =False)
r4['MinRank_desc']=r4['CountLength'].rank(method ='min',ascending =False)
r4['MeanRank_desc']=r4['CountLength'].rank(method ='average',ascending =False)
r4
```
