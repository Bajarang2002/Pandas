```python
import pandas as pd
import numpy as np
df = pd.read_csv("Top100moststreamed.csv")
df.head(10)
```

```python
# lambda apply function

res =df[['popularity']].apply(np.max)
res
```

```python
res =df[['popularity']].apply(lambda x: x.max())
res
```

```python
res =df[['popularity','energy','liveness']].apply(lambda x: x.max())
res
```

```python
res=df[['popularity','energy','liveness']].apply(lambda x: x.min())
res
```

```python
res=df[['popularity','energy','liveness']].apply(lambda x: x.max()-x.min())
res
```
