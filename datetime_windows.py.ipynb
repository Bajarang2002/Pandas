```python
import pandas as pd
import numpy as np
df = pd.read_csv("Electric_production.csv")
df.head(10)
```

```python
r1 =df.rename(columns ={"DATE":"Date","IPG2211A2N":"Value"})
r1.head(10)
```

```python
# Datetime function
r1['Date'] = pd.to_datetime(r1['Date'])
r1.dtypes

```

```python
r1['Year']=r1['Date'].dt.year
r1['Quarter']=r1['Date'].dt.quarter
r1['Month']=r1['Date'].dt.month
r1['Day']= r1['Date'].dt.day
r1['Weekday']=r1['Date'].dt.weekday
r1.head()
```

```python
r1['Year']=r1['Date'].apply(lambda x:x.year)
r1['Quarter']=r1['Date'].apply(lambda x:x.quarter)
r1['Month']=r1['Date'].apply(lambda x:x.month)
r1['Day']= r1['Date'].apply(lambda x: x.day)
r1['Weekday']=r1['Date'].apply(lambda x:x.weekday)
r1.head()
```

```python
r1 =df.rename(columns ={"DATE":"Date","IPG2211A2N":"Value"})
r1.head(10)
```

```python
r1 =r1[['Date','Value']]
r1
```

```python
r1 = r1 .set_index('Date')
r1
```

```python
# Window function
# shift

r1['Lag1']=r1['Value'].shift(1)
r1['Lag2']= r1['Value'].shift(2)
r1.head()
```

```python
# Difference

r1['Diff_m1']=r1['Value'].diff(1)
r1['Diff_m2']=r1['Value']-r1['Lag1']
r1['Diff_m3']= r1['Value'].diff(2)
r1['Diff_m4']= r1['Value']-r1['Lag2']
r1.head()
```

```python
# Rolling

r1['roll_m1'] = r1['Value'].rolling(2).mean()
r1['roll_m2'] = r1['Value'].rolling(3).mean()
r1.head()
```
